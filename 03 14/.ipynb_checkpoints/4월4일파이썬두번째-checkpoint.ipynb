# 데이터 정제하기 


```python
import numpy as np
```

```python
customerData = pd.read_csv("C:../dataset/customerdata.csv")
customerData.head()
```

```python
customerData["PRODUCTAGE_NEW"] = np.where(customerData["PRODUCTAGE"] < 1, 1,\
                                          np.where(customerData["PRODUCTAGE"] < 2, 2,\
                                          np.where(customerData["PRODUCTAGE"] < 3, 3, 5)))
                                        
```

```python
customerData.head()
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else: 
        return 5 
```

```python
customerData["PRODUCTAGE_NEW2"] = customerData.apply(refining, axis=1)
```

```python
customerData.head()
```

## PPT 29번 실습, 53주차 제거 및 YEAR WEEK 생성

```python
okselloutData = pd.read_csv("C:../dataset/kopo_channel_seasonality_new.csv")
okselloutData.head()
```

```python
okselloutData.dtypes
```

```python
okselloutData["YEARWEEK"] = okselloutData["YEARWEEK"].astype(str)  ## 데이터 타입 str로 바꿔줌
```

```python
okselloutData["YEAR"] = okselloutData["YEARWEEK"].str[0:4]
```

```python
okselloutData["WEEK"] = okselloutData["YEARWEEK"].str[4:]
```

```python
okselloutData.head()
```

```python
refinedSelloutData = okselloutData[okselloutData.WEEK.astype(int) <= 52]
```

```python
refinedSelloutData["WEEK"].max()
```

```python
import pandas as pd
selloutData = pd.read_csv('./Python_Fintech/Python_ST_EX/dataset/kopo_channel_seasonality_new.csv')
```

```python
selloutData.head()
```

```python
import numpy as np
selloutData['NEW_QTY'] = np.where(selloutData.QTY < 0, 0, selloutData.QTY)
```

```python
selloutData[selloutData.NEW_QTY < 0]
```

```python
selloutData.dtypes
```

```python
def customFunction(inputColumn):
    return inputColumn // 100
```

```python
selloutData['YEAR'] = customFunction(selloutData.YEARWEEK)
```

```python
selloutData['WEEK'] = selloutData.YEARWEEK % 100
```

```python
selloutData.head()
```

```python
refinedData = selloutData[selloutData.WEEK <= 52]
```

```python
refinedData.head()
```

```python
sortKey = ['REGIONID', 'PRODUCT', 'YEARWEEK']
sortedData = refinedData.sort_values(sortKey)
sortedData.head()
```

```python
sortedData = sortedData.reset_index(drop = True)
sortedData.head()
```

```python
sortedData['MA'] = sortedData.NEW_QTY.rolling(window = 15, center = True, min_periods = 1).mean()
sortedData
```

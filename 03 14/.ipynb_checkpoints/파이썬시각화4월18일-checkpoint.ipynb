```python
x = list(range(1,11,1))
```

```python
y = [i**2 for i in x]  ## 리스트안에 포문이 들어감 그 포문은 1에서 11까지 1씩 들어가는놈이 x에 담기는데 
                        ## 그걸 i**2 즉, 제곱한애들은 리스트에 담아줌
```

```python
y2 = [i**2.2 for i in x]  ## 포문의 디버깅..? 어케하냐?
```

```python
y3 = [i**2.5 for i in x]  
```

```python
plt.plot(x,y,"go--", label="first")
plt.plot(x,y2,"bo--", label="second")
plt.plot(x,y3,"ro--", label="test")
plt.legend(loc=0)  ## 좌측 상단에 라벨과 함꼐 보임 0 값이 베스트? 비어있는곳으로 감? 체크해볼것.
```

```python
plt.plot(x,y,"go--", label="first")
plt.plot(x,y2,"bo--", label="second")
plt.legend(loc=0)

plt.annotate('point',xy=(2,3), xytext=(4,4),arrowprops=dict(color="blue", width=2))  # 화살표 시작지점
```

```python
x=list(range(1,11,1))
y=[]
for i in x:
    y.append(i**2)
plt.plot(x,y)

# 차트 꾸미기
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정
plt.plot(x,y, 'bo', ) #차트 내 'o' 표기
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```

## 서브플롯 그리기

```python
# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 서브플롯
fig = plt.figure(figsize=(10,5))  ## figure 도화지의 사이즈는 10 과 5다
ax1 = fig.add_subplot(211)  ##   2행 1열에 1번째
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')  ## 쩜이 녹색인걸로 보아 디폴트값이 g 인듯,,? bo 로 바꾸면 파란색됨!!!
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(212)  ## ##   2행 1열에 2번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 서브플롯
fig = plt.figure(figsize=(10,5))  ## figure 도화지의 사이즈는 10 과 5다
ax1 = fig.add_subplot(121)  ##   1행 2얄에 1번째
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'bo')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(122)  ## ##   1행 2열에 2번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

## 이중축 활용하기

```python
# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 이중축
fig = plt.figure(figsize=(10,5))  ## 10,5 로 도화지를 그림
ax1 = fig.add_subplot(1,1,1)  ## 1행 1열에 하나 = 차트는 하나만 있으니 ㅇㅇ
plt.plot(x,y, 'b--',lw=1.5, label = '1st')
plt.plot(x,y, 'bo')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx()  ## twinx, x축과 같이 쓰겟다.
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

## PPT 15번 문제

```python
feature_Data = pd.read_csv("C:../dataset/feature_regression_example.csv")
```

```python
feature_Data_2015 = feature_Data[feature_Data.YEAR == 2015]
```

```python
x = feature_Data_2015[["WEEK"]] ## 대 괄호 1개랑 2개랑 데이터 형식이 다름 ,[[]]는 pacnas 형식, [] 는시리즈타입
```

```python
y = feature_Data_2015[["QTY"]]
```

```python
y2 = feature_Data_2015[["PRO_PERCENT"]]
```

```python
fig = plt.figure(figsize=(10,5))  ## 10,5 로 도화지를 그림
ax1 = fig.add_subplot(1,1,1)  ## 1행 1열에 하나 = 차트는 하나만 있으니 ㅇㅇ
plt.plot(x,y, 'bo--',lw=1.5,label='1st')
plt.grid(True)
plt.ylabel('qty') 
plt.title('subplot')

ax2 = ax1.twinx()  ## twinx, x축과 같이 쓰겟다. y축을 같이 쓴다고 안했으니 y축의 라벨값만 아래서 해주면됨
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
# plt.plot(x,y2, 'g')
# plt.xlabel('qty') ## x 라벨을 같이 쓰기 때문에 여기서 또 해줄필요 없음
plt.ylabel('promotion')
plt.legend(loc=0)
```

## 차트종류변경

```python
x=list(range(1,11,1))
y = [i**2 for i in x]

plt.figure(figsize=(7,5))  ## 캔버스 7,5 만들고~

plt.title('Scatter plt')
plt.scatter(x,y,marker='o', edgecolors='b') ## scatter 하면 아래와 같이찍힘
```

## 차원추가 (Scatter 차트)

```python
ok_kopo_selloutData = pd.read_csv("C:../dataset/kopo_channel_seasonality_new.csv")
```

```python
ok_kopo_selloutData.head()
```

```python
ok_kopo_selloutData["QTY_NEW"] = np.where(ok_kopo_selloutData["QTY"] < 0,0,ok_kopo_selloutData["QTY"])
ok_kopo_selloutData.head()
```

```python
ok_kopo_selloutData.head()
```

```python
ok_kopo_selloutData.dtypes
```

```python
ok_kopo_selloutData["YEARWEEK"] = ok_kopo_selloutData["YEARWEEK"].astype(str)
ok_kopo_selloutData.dtypes
```

```python
ok_kopo_selloutData["YEAR"] = ok_kopo_selloutData["YEARWEEK"].str[0:4]
```

```python
ok_kopo_selloutData["WEEK"] = ok_kopo_selloutData["YEARWEEK"].str[4:]
```

```python
ok_kopo_selloutData.head()
```

```python
refinedSelloutData = ok_kopo_selloutData[ok_kopo_selloutData.WEEK.astype(int) <=52 ]
```

```python
refinedSelloutData["WEEK"].max()
```

```python
refinedSelloutData.head()
```

```python
refinedSelloutData.head()
```

```python
grouDatago = refinedSelloutData.groupby(groupkey).agg(['mean','std'])['QTY_NEW']
```

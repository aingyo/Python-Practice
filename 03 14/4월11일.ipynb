## 1. 기본 이동평균 함수

```python
step1Data["MA"] = step1Data.NEW_QTY.rolling(window=15, center=True).mean()
```

```python
step1Data.head(10)
```

```python
step1Data["MA"] = step1Data.NEW_QTY.rolling(
    window=15, center=True, min_periods=1).mean()  
## window=15 고 center=True 이기 때문에 자기중심 기준으로 위 아래 7인데 min_periods=1 옵션때문에
## 자기 트루기준 아래 7과 자기 포함한 값의 평균값을 가짐
## min_periods=1 이라면 데이터가 1개라도 존재하면 평균을 낸다는거임
## window=5, center=False, min_periods=3 이라고 하면 center가 펄스고 window가 5인데 min_periods 가 3이기
## 때문에 3번째부터 평균이 구해짐

## 아래 참고
## https://ordo.tistory.com/67 
```

```python
def groupRolling(oneGroup):     ##  아래있는놈들 합쳐서 함수만듬
    
#     oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[2])
    indexGroupData = oneGroup.reset_index(drop=True)
    indexGroupData["MA2"] = indexGroupData["NEW_QTY"].rolling(window=5, center=True,min_periods=1).mean()
    
    return indexGroupData
```

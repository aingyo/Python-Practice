## PPT 26 복습

```python
import numpy as np
```

```python
customerdata = pd.read_csv("C:../dataset/customerdata.csv")
```

```python
customerdata.head()
```

```python
customerdata["PRODUCTAGE_NEW"] = np.where(customerdata["PRODUCTAGE"] < 1, 1, \
                                         np.where(customerdata["PRODUCTAGE"] < 2, 2, \
                                         np.where(customerdata["PRODUCTAGE"] < 3, 3, 5 )))
```

```python
customerdata.head()
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else: 
        return 5 
```

```python
customerdata["PRODUCTAGE_NEW"] = customerdata.apply(refining, axis=1)
```

```python
customerdata.head()
```

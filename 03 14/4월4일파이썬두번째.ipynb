# 데이터 정제하기 


```python
import numpy as np
```

```python
customerData = pd.read_csv("C:../dataset/customerdata.csv")
customerData.head()
```

```python
customerData["PRODUCTAGE_NEW"] = np.where(customerData["PRODUCTAGE"] < 1, 1,\
                                          np.where(customerData["PRODUCTAGE"] < 2, 2,\
                                          np.where(customerData["PRODUCTAGE"] < 3, 3, 5)))
                                        
```

```python
customerData.head()
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else: 
        return 5 
```

```python
customerData["PRODUCTAGE_NEW2"] = customerData.apply(refining, axis=1)
```
